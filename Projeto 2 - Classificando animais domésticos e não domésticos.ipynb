```python
#Importar biblioteca pandas responsável por importar o dataset
import pandas as pd
zooDs = pd.read_csv('C:/Users/Laisla Francisco/Documents/Python/zoo.csv')
zooDs.head()
```

```python
#Excluir coluna com strings
zooDs.drop('animal_name', axis = 1, inplace=True)
zooDs.head()
```

```python
#Separar variáveis preditoras e a variável alvo (domestic)
y = zooDs['domestic']
x = zooDs.drop('domestic', axis = 1)
```

```python
#Separar dados para treino e para teste
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size = 0.3)

#Visualizar qtd de dados de treino e teste
print('Treino: ', x_treino.shape)
print('Teste: ',x_teste.shape)
```

```python
#Importar algoritmo de decisão
from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(x_treino,y_treino)

#Exibe o quanto o algoritmo acertou dos dados inseridos
resultado = modelo.score(x_teste, y_teste)
print('Acurácia: ', resultado)
```
